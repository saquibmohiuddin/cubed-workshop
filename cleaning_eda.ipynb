```python
pop_df.drop(['LA Code (2018 boundaries)', 'LA name (2018 boundaries)', 
             'LA Code (2021 boundaries)', 'LA name (2021 boundaries)'], 
            axis=1, inplace=True, errors='ignore')
```

```python
crime_df = pd.read_csv('clean_data\clean_data_v2.csv')
crime_df
```

```python
crime_df.drop(['Crime ID', 'Reported by'], axis=1, inplace=True, errors='ignore')
```

```python
crime_df
```

```python
crime_df['Outcome type'].unique()
```

```python
crime_df['Conviction'] = crime_df['Outcome type'].apply(lambda x: 'prosecuted' if 'charged' in x else 'not-prosecuted')
```

```python
crime_df
```

```python
pop_df
```

```python
lsoa_crime_df = crime_df.groupby(['LSOA code', 'LSOA name']).agg({'Month':'count',
                                            'Falls within':', '.join,
                                            'Falls within':'unique',
                                            'Location':', '.join,
                                            'Location':'unique',
                                            'Crime type':', '.join,
                                            'Crime type':'unique',
                                            'Conviction':', '.join}).reset_index()
```

```python
lsoa_crime_df
```

```python
def word_count(str):
    counts = {}
    words = str.split(', ')
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts
```

```python
word_count('prosecuted, not-prosecuted, not-prosecuted, not-prosecuted')
```

```python
lsoa_crime_df['Conviction'] = lsoa_crime_df['Conviction'].apply(word_count)
```

```python
lsoa_crime_df
```

```python
col = lsoa_crime_df['Conviction'].apply(lambda x: pd.Series(x))
```

```python
col
```

```python
df = pd.concat([lsoa_crime_df, col], axis=1)
df
```

```python
df[['Falls within', 'Location', 'Crime type']] = df[['Falls within', 'Location', 'Crime type']].applymap(lambda x: ', '.join(x))
df
```

```python
pop_crime_location_df = df.merge(pop_df, left_on=['LSOA code', 'LSOA name'], right_on=['LSOA Code', 'LSOA Name'], how='left')
pop_crime_location_df
```

```python
pop_crime_location_df.rename({'Month':'Crime Count'}, axis=1, inplace=True, errors='ignore')
```

```python
pop_crime_location_df = pop_crime_location_df[['LSOA code', 'LSOA name', 'Crime Count', 'Crime type', 'Location', 'Falls within', 'All Ages', 'not-prosecuted', 'prosecuted']]
pop_crime_location_df
```

```python
import re
pop_crime_location_df['Location'] = pop_crime_location_df['Location'].apply(lambda x: re.sub("On or near ", "", x))
```

```python
from datetime import datetime


def save_file(dataframe):
    current_datetime = datetime.now().strftime('%Y%m%d_%H%M%S')
    file_name = 'pop_crime'
    output_file = f'clean_data/{file_name}_{current_datetime}.csv'
    dataframe.to_csv(output_file, index=False)
    
    return output_file
```

```python
save_file(pop_crime_location_df)
```
